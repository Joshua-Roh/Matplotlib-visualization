```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Gas Price Data

```python
df_gas = pd.read_csv('gas_prices.csv')
df_gas
```

### Line Graph

```python
plt.figure(figsize = (8,4))

plt.plot(df_gas['Year'], df_gas['USA'], 'r.-', label = 'United States')
plt.plot(df_gas['Year'], df_gas['Canada'], 'b.-', label = 'Canada')
plt.plot(df_gas['Year'], df_gas['Australia'], 'g.-', label = 'Australia')


#Another way to select the specified countries and their gas prices
#countries_to_look_at = ['USA', 'Canada', 'Mexico', 'South Korea']
#for country in df_gas:
#   if country in countries_to_look_at:
#       plt.plot(df_gas['Year'], df_gas[country], marker = '.', label = country)



plt.title('Gas Prices over Time(in USD)', fontdict = {'fontname': 'Comic Sans MS', 'fontweight': 'bold', 'fontsize': 16})
plt.xlabel('Year', fontdict = {'fontname': 'Comic Sans MS'})
plt.ylabel('Price', fontdict = {'fontname': 'Comic Sans MS'})

plt.legend()
plt.xticks(df_gas.Year[::3])

plt.savefig('GasPrices.png')

plt.show()
```

## Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

### Histogram

```python
bins = [40, 50, 60, 70, 80, 90, 100]

plt.figure(figsize = (6,5))

plt.title('Distribution of Skill Level in FIFA 2018')

plt.hist(fifa.Overall, bins = bins)

plt.xticks(bins)
plt.xlabel('Skill Level')
plt.ylabel('Number of Players')


plt.savefig('Hist_Skill_Level.png')

plt.show()
```

### Pie Chart #1

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left, right], labels = ['Left Footed Players', 'Right Footed Players'], colors = ['#abcdef', '#aabbcc'], autopct='%.2f %%')
plt.title('Foot Preference of Fifa Players')

plt.savefig('FIFA_Foot_Preference.png')

plt.show()
```

### Pie Chart #2

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

light = fifa.loc[fifa['Weight'] < 125].count()[0]
light_medium = fifa.loc[(fifa['Weight'] >= 125) & (fifa['Weight'] < 150)].count()[0]
medium = fifa.loc[(fifa['Weight'] >= 150) & (fifa['Weight'] < 175)].count()[0]
medium_heavy = fifa.loc[(fifa['Weight'] >=175) & (fifa['Weight'] < 200)].count()[0]
heavy = fifa.loc[fifa['Weight'] >=200].count()[0]
```

```python
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = [0.4, 0.1, 0, 0, 0.4]

plt.figure(figsize = (8, 8))
plt.style.use('ggplot')
plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)
plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.savefig('FIFA_Weight_Distribution.png')

plt.show()
```

### Box and Whiskers Plot

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
juventus = fifa.loc[fifa.Club == 'Juventus']['Overall']
psg = fifa.loc[fifa.Club == 'Paris Saint-Germain']['Overall']
manchester = fifa.loc[fifa.Club == 'Manchester United']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']

labels = ['Barcelona', 'Juventus', 'PSG', 'MU', 'Real Madrid']

plt.figure(figsize = (5,8))

boxes = plt.boxplot([barcelona, juventus, psg, manchester, madrid], labels = labels)
plt.ylabel('FIFA Overall Rating')
plt.title('Professional Football Team Comparison')


plt.savefig('Whisker_FIFA_Team_Comparison.png')
plt.show()
```
